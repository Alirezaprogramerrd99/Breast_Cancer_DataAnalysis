```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

###  Breast Cancer data set configirations: 

```python
df = pd.read_csv('dataR2.csv')   # loading the data set in to the pandas dataframe.
df
```

```python
#collecting data set index for x-axis.
indexes = list(df.index)

# extracting class label values of the problem
Ylabels = df['Classification']

# extracting features of the problem. (deleting the class label values)
features = df.drop('Classification', axis = 1)   
```

```python
def extractFeaturesValues(feature_name):
    return list(features[feature_name])
def extractlstYlabels():
    return list(Ylabels)
```

```python
features.describe()
```

```python
df_dummy = df.copy(deep = True)
df_dummy['Classification'].replace({1:'healthy', 2: 'breast cancer'}, inplace = True)
df_dummy.groupby('Classification').mean()
```

###  Ploting features vs indexes according class labels:
#### for BMI:

```python
#collecting BMI values from data set for y-axis.
bmi_lst = extractFeaturesValues('BMI') 
colors = ['red', 'green']
#plt.scatter(indexes, bmi_lst, c = extractlstYlabels(), cmap = matplotlib.colors.ListedColormap(colors))

s = plt.scatter(indexes, 
            bmi_lst,
            s = 40,
            c = Ylabels.astype('category').cat.codes, cmap = matplotlib.colors.ListedColormap(colors))

plt.xlabel('indexes')
plt.ylabel('BMI')
plt.title('plot for BMI vs indexes according to class-labels')
plt.legend(handles=s.legend_elements()[0],
           labels=['healthy 1', 'breast cancer 2'],
           title="labels")
plt.show()
```

 ####  for all data set features:

```python
column_lst = list(features.columns)

for col in column_lst: 
    
    featureName_lst = extractFeaturesValues(col)
    scatter = plt.scatter(indexes, featureName_lst, s = 35, c = Ylabels.astype('category').cat.codes, cmap = matplotlib.colors.ListedColormap(colors))
    plt.xlabel('indexes')
    plt.ylabel(col)
    plt.title("".join (["plot for ", col, " vs indexes according to class-labels"]))
    plt.legend(handles = scatter.legend_elements()[0], labels = ['healthy 1', 'breast cancer 2'], title = "labels")
    plt.show()
    
```

### Ploting features vs features according class labels:

```python
def plotFigFeatures(features):
    f1_lst = extractFeaturesValues(features[0])
    f2_lst = extractFeaturesValues(features[1])

    s = plt.scatter(f1_lst, f2_lst, s = 40, c = Ylabels.astype('category').cat.codes, cmap = matplotlib.colors.ListedColormap(colors))

    plt.xlabel(features[0])
    plt.ylabel(features[1])
    plt.legend(handles=s.legend_elements()[0], labels=['healthy 1', 'breast cancer 2'], title="labels")
    plt.show()
```

```python
# we want to plot below features
f_plotLst = [('Insulin', 'BMI'), ('HOMA', 'Insulin'), ('Leptin', 'MCP.1'), ('Age', 'BMI')]   # format: (x-axis, y-axis)

for t in f_plotLst:
    plotFigFeatures(t)
```
